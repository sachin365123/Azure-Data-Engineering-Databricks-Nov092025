✅ 2. From the Databricks UI

Go to the Catalog tab (left sidebar in the Databricks workspace).

With UC enabled → you’ll see a Catalogs view at the top (each catalog as a separate container).

Without UC → you’ll just see Databases (schemas) under the Hive metastore.

✅ 3. Cluster permissions

UC requires:

A Unity Catalog–enabled cluster (set when you create the cluster).

Proper permissions (USE CATALOG, USE SCHEMA, etc.) to access catalogs.

If your cluster type dropdown doesn’t mention Unity Catalog when creating, your workspace probably isn’t UC-enabled.

### Solution 
Creating a UC-enabled cluster (you’ll see “Access mode: Standard (Shared),Dedicated (Single user), or No isolation” options, not [Legacy] Single USer Standard).
